```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
books = pd.read_csv('BX-Books.csv', sep=';', error_bad_lines=False, encoding="latin-1")
book_ratings = pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users_info = pd.read_csv('BX-Users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
```

```python
 books.head()
```

```python
books = books.drop(columns = ['Year-Of-Publication', 'Publisher','Image-URL-S', 'Image-URL-M', 'Image-URL-L','Book-Author'])
books.head()
```

```python
book_ratings.head()
```

```python
users_info.head()
```

```python
users_info['Country'] = users_info['Location'].str.split(',').str[-1].str.strip()
```

```python
users_info = users_info.drop(columns = ['Location', 'Age'])
```

```python
# Countries = users_info['Country'].unique()
# Countries.sort()
```

```python
#Countries
```

```python
usa_strings = ['good old u.s.a.', 'good old usa !','america','u.s. of a.', 'u.s. virgin islands', 'u.s.a!', 'u.s.a.', 'u.s.a>', 'u.s>','unite states', 'united sates', 'united staes', 'united state', 'united statea', 'united stated', 'united stated of america','united states','united states of america', 'us', 'usa', 'usa & canada', 'usa (currently living in england)', 'usa"', 'ussurfing', 'ust', 'uusa','wonderful usa']
india_strings = ['in','india','india"','indiai']
users_info['Country'] = users_info['Country'].replace(to_replace = usa_strings, value='usa')
users_info['Country'] = users_info['Country'].replace(to_replace = india_strings, value='india')
```

```python
users_info.head()
```

```python
combined_book_user_rating = pd.merge(book_ratings, books, how="left", on="ISBN")
combined_book_user_rating = pd.merge(combined_book_user_rating, users_info, how="left",on="User-ID")
```

```python
#combined_book_user_rating = combined_book_user_rating.loc[combined_book_user_rating['Country']=='usa']
```

```python
#combined_book_rating = combined_book_user_rating.drop(columns = ['Country', 'Book-Title'])
```

```python
combined_book_user_rating
```

```python
combined_total_ratings = combined_book_user_rating.groupby('ISBN').count()['Book-Rating']
```

```python
combined_total_ratings = combined_total_ratings.reset_index()
combined_total_ratings = combined_total_ratings.rename({'Book-Rating':'totalNoOfRatings'}, axis='columns')
```

```python
combined_total_ratings.columns
```

```python
combined_total_ratings_final = combined_total_ratings.merge(book_ratings, how="right", on="ISBN")
combined_total_ratings_final = combined_total_ratings_final.merge(users_info, how="right", on="User-ID")
```

```python
combined_total_ratings_final['totalNoOfRatings'].describe()
```

```python
combined_total_ratings_final.head()
```

```python
popular_books_df = combined_total_ratings_final[combined_total_ratings_final['totalNoOfRatings']>=500] 
```

```python
popular_books_df_usa = popular_books_df.loc[popular_books_df['Country']=='usa']
```

```python
popular_books_df_usa = popular_books_df_usa.drop(columns = ['Country', 'totalNoOfRatings'])
```
